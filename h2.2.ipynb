```python
import pandas as pd
path = "2017_jun_final.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.drop(["Специализация","Университет","Предметная.область","Валюта","cls"], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
python_data = df.loc[df["Язык.программирования"] == "Python",:]
```

```python
python_data.shape
python_data.head()
```

```python
python_df_2 = python_data.groupby(["Должность"])
```

```python
python_df_2.agg({"Зарплата.в.месяц": ["min","max"]})
```

```python

def fill_avg_salary(salary):
    avg = python_data['Зарплата.в.месяц'].mean()
    return avg

python_data['avg'] = python_data['Зарплата.в.месяц'].apply(fill_avg_salary)

```

```python
python_data['avg'].describe()
```

```python
python_data.to_csv('data_with_avg.csv')
```

```python
import matplotlib.pyplot as plt
data = df["Язык.программирования"].value_counts()
labels = data.index
plt.figure(figsize=(24, 24))
plt.pie(data, labels=labels, autopct="%.2f%%")
plt.show()
```

```python
data = df["Город"].value_counts()
x = data.index
plt.figure(figsize=(40, 10))
plt.bar(x,data)
plt.xlabel('Місто')
plt.ylabel('Кількість працівників')
plt.show()
```
